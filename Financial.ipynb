3. #### Columns
**Segment**: The market segment in which the sales occurred (e.g., Government, Midmarket).

**Country**: The country where the sales transaction took place.

**Product**: The specific product sold during the transaction.

**Discount Band**: The discount category applied to the product.

**Units Sold**: The number of units sold in each transaction.

**Manufacturing Price**: The cost to manufacture one unit of the product.

**Sale Price**: The selling price per unit of the product.

**Gross Sales**: The total sales amount before applying any discounts.

**Discounts**: The total discounts applied to the gross sales.

**Sales**: The net sales amount after discounts.

**COGS (Cost of Goods Sold)**: The direct costs attributed to the production of the goods sold.

**Profit**: The profit earned from the sales, calculated as Sales minus COGS.

**Date**: The date on which the transaction occurred.

**Month Number**: The numeric representation of the month of the transaction.

**Month Name**: The name of the month in which the transaction occurred.

**Year**: The year in which the transaction occurred.

##### Converting `Discount` and `Profit` columns

- `Discounts` column contains strings that cannot be directly converted to numeric values, such as " - " (possibly representing missing values). To handle this, we should first replace these non-numeric strings with NaN values and then convert the column to numeric.

```python
# 7. Clean Discounts column
# Step 1: Replace problematic strings with NaN
df[' Discounts '] = df[' Discounts '].replace(' -   ', np.nan)

# Step 2: Remove unwanted characters
df[' Discounts '] = df[' Discounts '].replace('[\$,]', '', regex=True)

# Step 3: Convert to numeric
df[' Discounts '] = pd.to_numeric(df[' Discounts '])

# Handle NaN values (e.g., fill with 0 or drop)
df[' Discounts '] = df[' Discounts '].fillna(0)

# Display the cleaned Discounts column
df[' Discounts '].head()
```

- `Profit` column contains strings that cannot be directly converted to numeric values, such as " - ". Similar to the Discounts column, we need to handle these non-numeric strings appropriately.

- `Profit` column also contains parentheses (e.g., "(4533.75)"). To handle these values, we need to account for the parentheses and convert them to negative numbers.

```python
# 8. Clean Profit column
# Step 1: Remove unwanted characters
df[' Profit '] = df[' Profit '].str.replace('[\$,]', '', regex=True)

# Step 2: Convert to numeric
df[' Profit '] = pd.to_numeric(df[' Profit '], errors='coerce')

# Handle NaN values (e.g., fill with 0 or drop)
df[' Profit '] = df[' Profit '].fillna(0)
```
